## Exercise 2

Write your salary computation to give an employee 1.5 times the hourly rate for hours worked above 40 hours.


```python
hrs = float(input("Enter Hours: ")) #prompt the user to enter hours
r = float(input("Enter Rate: ")) #prompt the user to enter rate

if hrs >= 40:
    pay = (40 * r) + ((hrs - 40) * r * 1.5) #if hours is greater than or equal to 40 calculate pay with overtime hours at 1.5 times the rate
else:
    pay == hrs * r #if its not greater than or equal to 40 hours calculate pay for all hours at the regular rate

print("Pay:", pay) #display calculated pay
```

# CHALLENGE

Write a program to prompt for a score between 0 and 100 and print a grade using the following table:

Score           Grade
[ 97, 100]      1.0
[ 93,  97)        1.25
[ 89,  93)        1.5
[ 85,  89)        1.75
[ 81,  85)        2.0
[ 77,  81)        2.25
[ 74,  77)        2.5
[ 0,   74)        3.0

```python
totalScore = float(input("Enter a score between 0 and 100: ")) #prompts user to enter score

if totalScore >= 97: #if score is greater than or equal to 97, the grade is 1.0
    grade = 1.0
elif totalScore >= 93: #if score is greater than or equal to 93, the grade is 1.25
    grade = 1.25
elif totalScore >= 89: #if score is greater than or equal to 89, the grade is 1.5
    grade = 1.5
elif totalScore >= 85: #if score is greater than or equal to 85, the grade is 1.75
    grade = 1.75
elif totalScore >= 81: #if score is greater than or equal to 81, the grade is 2.0
    grade = 2.0
elif totalScore >= 77: #if score is greater than or equal to 77, the grade is 2.25
    grade = 2.25
elif totalScore >= 74: #if score is greater than or equal to 74, the grade is 2.5
    grade = 2.5
else:
    grade = 3.0 #if score is less than 74, the grade is 3.0

print(f"The grade for the score {totalScore} is {grade}.") #displays the grade for score prompt by user
```
