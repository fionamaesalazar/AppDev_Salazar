### Exercise 1

Rewrite your pay computation with time-and-a-half for overtime and create a function called compute pay which takes two parameters (hours and rate).

```python
def compute_Pay(hours, rate):
  if hours >= 40:
    overTime_hrs = hours - 40
    overTime_pay = overTime_hrs * (rate * 1.5)
    reg_Pay = 40 * rate
    return  reg_Pay +  overTime_pay
  else:
     return hours * rate
hours = float(input("Enter Hours: ")) #prompt the user to enter hours
rate = float(input("Enter Rate: ")) #prompt the user to enter rate

pay = compute_Pay(hours,rate)
print(f"Pay: {pay}")
```

# CHALLENGE

Write a function called computegrade that takes a score as its parameter and returns a grade as a string using the following table:

    Score           Grade
    [ 97, 100]      1.0
    [ 93,  97)		1.25
    [ 89,  93)		1.5
    [ 85,  89)		1.75
    [ 81,  85)		2.0
    [ 77,  81)		2.25
    [ 74,  77)		2.5
    [ 0,   74)		3.0

    Run the program repeatedly to test the various different values for input.

```python
def computegrade(score):
    if score >= 97:
        return "1.0"
    elif score >= 93:
        return "1.25"
    elif score >= 89:
        return "1.5"
    elif score >= 85:
        return "1.75"
    elif score >= 81:
        return "2.0"
    elif score >= 77:
        return "2.25"
    elif score >= 74:
        return "2.5"
    else:
        return "3.0"

while True:
    try:
        score = float(input("Enter a score between 0 and 100: "))
        if 0 <= score <= 100:
            grade = computegrade(score)
            print(f"The grade for the score {score} is {grade}.")
            break
        else:
            print("Score must be between 0 and 100.")
    except ValueError:
        print("Invalid input. Please enter a number.")
```
