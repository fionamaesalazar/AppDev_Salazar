### Exercise #2

Write a program to prompt the user for hours and rate per hour to compute gross pay.

```python
hr = input('Enter Hour:') #we initialized a variable hr
rate = input('Enter Rate:') #we initialize the variable rate
pay = float(hr) * float(rate) #we first initialized the variable pay the proceed with the calculation
print('Pay:', pay) #we display the computed gross pay
```

# CHALLENGE

Write a program which prompts the user for a Celsius temperature, convert the temperature to Fahrenheit, and print out the converted temperature.

```python
c = float(input("Enter Temperature in Celsius: "))
fah = (c * 9/5) + 32
print (f"Celsius {c} converted to Fahrenheit is {fah}")
```
