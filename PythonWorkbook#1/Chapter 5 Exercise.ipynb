Exercise 1:

Investigate on the difference between `is` and `==` given the following variables.
```python
a = 2
b = 2.0
c = a
```


````python
a = 2
b = 2.0
c = a

#The difference of is and == is that:
#is - doing object identity checks. If two variables point to the same memory location, it returns True, if not, it returns False.
# == indicates that ths values are equal. If the values of the two variables are equal, it returns True, if not, it returns False.

print("a is b:", a is b)  # a is b would be False since they are different objects in memory, even though their values are the same.
print("a == b:", a == b)  # a == b is True as it compares the values, which are equal 2 == 2.0
print("a is c:", a is c)  # a is c is True as they refer to the same object, both point to the integer 2
print("a == c:", a == c)  # a == c is True since it compares the values, which are equal 2 == 2


````

### CHALLENGE 1:

Write a program which repeatedly reads numbers until the user enters "done". Once "done" is entered, print out the total, count, and average of the numbers. If the user enters anything other than a number, detect their mistake using try and except and print an error message and skip to the next number.

    Enter a number: 4
    Enter a number: 5
    Enter a number: bad data
    Invalid input
    Enter a number: 7
    Enter a number: done
    16 3 5.333333333333333
    
[PEW] E5.1

````python
total = 0
count = 0
while True:
    no_ = input("Enter a number: ")
    if no_ == "done":
        break
    try:
        total += float(no_)
        count += 1
    except ValueError:
        print("Invalid input")

if count:
    print(total, count, total / count)
else:
    print("No valid numbers were entered.")
````

### CHALLENGE 2:


Write another program that prompts for a list of numbers as above and at the end prints out both the maximum and minimum of the numbers instead of the average.

````python
total = 0
count = 0
max_num = None
min_num = None

while True:
    num = input("Enter a number: ")
    if num == "done": #type done to finish
        break
    try:
        num = float(num)
        total += num
        count += 1
        if max_num is None or num > max_num:
            max_num = num
        if min_num is None or num < min_num:
            min_num = num
    except ValueError:
        print("Invalid input. Please enter a number.")
        continue

if count:
    print("Maximum number:", max_num)
    print("Minimum number:", min_num)
    print("Sum:", total)
    print("Average:", total / count)
else:
    print("No valid numbers were entered.")
````
